```python
import matplotlib.pyplot as plt
%matplotlib inline
from pycalphad import Database, binplot
import pycalphad.variables as v
```

```python
file = 'Ag_Cu.tdb'

# load database
dbf = Database(file)

# Define the components in our alloy
comps = ['Ag','Cu','VA']

# Get all possible phases progammatically
phases = dbf.phases.keys()

# Pressure [=] Pa
# Temperature [=] K
# X == Composition/mol percent/mass percent
binplot(dbf, comps, phases, {v.N:1, v.P:101325, v.T: (473,1473,10), v.X('Cu'):(0,1,0.02)})

# plot it!
plt.show()
                    
```

```python
# Now, let's do Ternary Phase Diagrams:
from pycalphad import ternplot
```

```python
file = 'Al_Mg_Si.tdb'

dbf = Database(file)
comps = ['AL','MG','SI','VA']

phases = list(dbf.phases.keys())
conds = {v.T: 973, v.P: 101325, v.X('AL'):(0,1,0.015), v.X('MG'):(0,1,0.015)}

ternplot(dbf, comps, phases, conds, x = v.X('AL'), y = v.X('MG'))
plt.show
```
