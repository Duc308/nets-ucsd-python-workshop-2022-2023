```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('PECVD Randomly Generated Data.csv')
data
```

```python
data = data.drop_duplicates(ignore_index=True) #delete all duplicates


x = data.drop(['k value'], axis=1)
y = data['k value']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
linear_model = LinearRegression() # Create linear regression object
linear_model.fit(X_train, y_train)
y_prediction = linear_model.predict(X_test)
mae = mean_absolute_error(y_test,y_prediction)
print(f'The Mean Absolute Error is {mae}')
```

```python
import pickle

model_filename = 'Predicting_k_value_model.pkl'
pickle.dump(linear_model, open(model_filename, 'wb'))
```
